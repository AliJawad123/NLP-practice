```python
import pandas as pd
```

```python
df=pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

```python
df = df.iloc[:, 0:2]
```

```python
df
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
df
```

# Now spliting the data into test and train dataset

```python
from sklearn.model_selection import train_test_split


# Here the message is the X and spam is Y

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

# loading the countvectorizer for Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()


# Here inside the transform when we supply the x train it will generate the BOW for the X train

X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
X_trn_BOW=X_train_cv.toarray() # this is the matrix for the BOW
```

```python
X_trn_BOW[1]
```

```python
X_trn_BOW.shape # here the output tells us that we have total 7770 words in our BOW and 4457 emails
```

```python
v.get_feature_names_out()[7000:7700]
```

```python
dir(v)
```

```python
v.vocabulary_ 
```

# Now we will train our model

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)  # here we have trained our model with the BOW of X train and Y train
```

```python
X_test_cv = v.transform(X_test) # to predict the output on the X_test we will change it into X test BOW
```

```python
y_pred = model.predict(X_test_cv)  # Now we will make our prediction
```

# Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
